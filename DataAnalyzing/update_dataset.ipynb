```python
import pandas as pd
import numpy as np
import os
```

# Combine all the weather files

```python
def name(path):
    file_name = []
    a = os.listdir(path)
    # a.sort()
    print(a)
    for j in a:
        if os.path.splitext(j)[1] == '.csv':
            file_name.append(j)
    return file_name
```

```python
def dir_process(path, dirpath):
    file_name = name(path)
    df = pd.DataFrame()
    for fn in file_name:
        df = data_process(df, path + fn, dirpath)
    df.to_csv(dirpath)
    return
```

```python
def data_process(df, path, dirpath):
    data = pd.read_csv(path)
    df = pd.concat([df, data])
    return df
```

```python
dir_process('./0_655/', './weather.csv')
```

# Merge flight, weather, airport tables

```python
weather = pd.read_csv('./weather.csv')
airport = pd.read_csv('./airport_cleaner.csv')
weather_airport = pd.merge(weather, airport, how='inner', on=['latitude', 'longitude']).drop(['Unnamed: 0_x','Unnamed: 0_y'], axis=1)
weather_airport.head(10)
```

```python
flight = pd.read_csv('./winter_2021_clean.csv')
```

```python
flight_weather = pd.merge(flight, weather_airport, how='inner', left_on=['flight_date', 'departure_airport'], 
                          right_on=['datetime', 'airport_name'])
flight_weather.head(10)
```

```python
flight_weather = flight_weather.iloc[:, :-11]
```

```python
flight_weather = pd.merge(flight_weather.loc[:, :], weather_airport, how='inner', left_on=['flight_date', 'arrival_airport'], 
                          right_on=['datetime', 'airport_name'], suffixes=('_departure', '_arrival'))
flight_weather.drop(flight_weather[-12:],axis=1)
flight_weather.head(10)
```

```python
flight_weather = flight_weather.iloc[:, :-11]
```

```python
flight_weather.loc[:, ['departure_iata', 'arrival_iata']]
```

```python
flight_weather = flight_weather.drop(['iata_code_x_departure', 'iata_code_x_arrival'], axis=1)
```

```python
# export final csv
flight_weather.to_csv('./flight_weather.csv')
```
