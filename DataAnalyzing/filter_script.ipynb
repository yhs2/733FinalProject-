```python
import pandas as pd


flight_df = pd.read_csv('./winter_2021_clean.csv')

departure_df = flight_df.groupby('departure_icao').agg(num_flights = ('departure_iata','count')).sort_values(by = 'num_flights').reset_index()

arrival_df = flight_df.groupby('arrival_icao').agg(num_flights = ('departure_iata','count')).sort_values(by = 'num_flights').reset_index()

# departure_df.merge(arrival_df, )
# departure_df
res = departure_df.merge(arrival_df, left_on='departure_icao', right_on='arrival_icao')
res['flight_all'] = res['num_flights_x'] + res['num_flights_y']
res = res.sort_values('flight_all')
res.tail(70)

res = res['departure_icao'].tail(80).to_list()
len(res)
```

```python
airport_df = pd.read_csv('airport_clean.csv')
# arrival_df[arrival_df['arrival_icao']]
overall_res = airport_df[airport_df['icao_code'].isin(res)]
# res.to_csv('|airport_important.csv')
```

```python
overall_res.to_csv('airport_important.csv')
```

```python
departure_iata_list =list(flight_df.departure_iata.value_counts().index)
arrival_iata_list = list(flight_df.arrival_iata.value_counts().index)
```

```python
overall_iata_list  = set(departure_iata_list + arrival_iata_list)
len(overall_iata_list)
```

```python
airport_df = pd.read_csv('./airport_clean.csv')
airport_df
```

```python
res = airport_df[airport_df['iata_code'].isin(overall_iata_list)]
res.drop(res.columns[0], axis=1,inplace=True)
```

```python
res.to_csv('airport_cleaner.csv')
```
