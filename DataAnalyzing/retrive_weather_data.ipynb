```python
import requests
import pandas as pd
from datetime import date, timedelta
import numpy as np
```

```python
airport_df = pd.read_csv('./airport_cleaner.csv')
airport_df.shape
```

```python

data = []

for index, row in airport_df.iterrows():
    # 0 - 655
    # 656 - 1311
    # 1311- 1967
    # 1967 - 2622
    # 2622 - 3044
    if index>=0 and index<=163:
        one_data = []
        lat = row['latitude']
        lon = row['longitude']
        iata_code = row['iata_code']
        params = {
            'key': 'd9aedb8fc45342f7ac2d000204861195',
            'start_date': '2021-12-01',
            'end_date': '2022-03-01',
            'lat':  str(lat),
            'lon':  str(lon),


        }
        api_result = requests.get('https://api.weatherbit.io/v2.0/history/daily', params)
        print(index)
        try:
            if 'data' in api_result.json():
                api_response = api_result.json()['data']
                for cur_date in api_response:
                    part_df = pd.json_normalize(cur_date)
                    one_data.append(part_df)

                one_df = pd.concat(one_data)
                one_df['iata_code']=iata_code
                one_df['latitude'] = lat
                one_df['longitude'] = lon
                data.append(one_df)
            else:
                print(lat,lon,iata_code)
                print(api_result.json())
        except ValueError:  # includes simplejson.decoder.JSONDecodeError
            print(api_result)
            print(lat,lon,iata_code)
df = pd.concat(data)

df.to_csv('./weather_0_163.csv',index=False)
```
